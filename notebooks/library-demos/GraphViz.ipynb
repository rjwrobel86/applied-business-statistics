# Diagrams / Flowcharts w/ GraphViz

```python
from graphviz import Digraph

dot = Digraph("Hypothesis_Test_Chooser", format="svg")

#Layout
dot.attr(
    rankdir="TB",
    splines="ortho",
    nodesep="0.70",
    ranksep="0.95",
    pad="0.25",
    bgcolor="white",
    concentrate="false"
)

#Styles
dot.attr("node", shape="box", style="rounded", fontsize="12", fontname="Helvetica", margin="0.18")
dot.attr("edge", fontsize="10", fontname="Helvetica", arrowsize="0.8")

dot.node("START", "START")

dot.node(
    "OUTCOME",
    "Outcome type?\n"
    "• Means\n"
    "• Proportions\n"
    "• Categorical\n"
    "• Quant–Quant"
)

dot.edge("START", "OUTCOME")

dot.node("H_MEAN", "Means", shape="box", style="rounded,bold")
dot.node("H_PROP", "Proportions", shape="box", style="rounded,bold")
dot.node("H_CAT", "Cat×Cat", shape="box", style="rounded,bold")
dot.node("H_QQ", "Quant–Quant", shape="box", style="rounded,bold")

dot.edge("OUTCOME", "H_MEAN")
dot.edge("OUTCOME", "H_PROP")
dot.edge("OUTCOME", "H_CAT")
dot.edge("OUTCOME", "H_QQ")

#Force the 4 headers to sit on the same horizontal row
with dot.subgraph() as s:
    s.attr(rank="same")
    s.node("H_MEAN")
    s.node("H_PROP")
    s.node("H_CAT")
    s.node("H_QQ")

dot.node(
    "M_GATE",
    "Check conditions (means)\n"
    "• Random/independent\n"
    "• If n is small: roughly normal (or no extreme outliers)\n"
    "• If n is large: CLT often ok"
)
dot.edge("H_MEAN", "M_GATE")

with dot.subgraph(name="cluster_mean") as m:
    m.attr(label="Means (quantitative outcomes)", fontsize="12", fontname="Helvetica", margin="18")

    m.node("M_GROUPS", "How many groups/samples?\n• 1\n• 2\n• 3+")
    m.edge("M_GROUPS", "M1_DEC", label="1")
    m.edge("M_GROUPS", "M2", label="2")
    m.edge("M_GROUPS", "M3", label="3+")

    m.node("M1_DEC", "1 sample:\nIs population σ known?\n(rare in practice)")
    m.node("M1Z", "One-sample z test for mean")
    m.node("M1T", "One-sample t test for mean")
    m.edge("M1_DEC", "M1Z", label="Yes")
    m.edge("M1_DEC", "M1T", label="No / unknown")

    m.node("M2", "2 samples:\nPaired or independent?")
    m.node("PAIRED", "Paired:\nPaired t test\n(test mean of differences)")
    m.node("INDEP", "Independent:\nWelch two-sample t test\n(default)")
    m.edge("M2", "PAIRED", label="Paired")
    m.edge("M2", "INDEP", label="Independent")


    m.node("M3", "3+ groups:\nOne-way ANOVA")

dot.edge("M_GATE", "M_GROUPS")


dot.node(
    "P_GATE",
    "Check conditions (proportions)\n"
    "• Random/independent\n"
    "• Success-failure:\n"
    "  expected successes ≥ 10 AND expected failures ≥ 10"
)
dot.edge("H_PROP", "P_GATE")

with dot.subgraph(name="cluster_prop") as p:
    p.attr(label="Proportions (binary outcomes)", fontsize="12", fontname="Helvetica", margin="18")

    p.node("P_GROUPS", "How many groups?\n• 1\n• 2\n• 3+")
    p.node("P1", "1 proportion:\nOne-proportion z test")
    p.node("P2", "2 proportions:\nTwo-proportion z test")
    p.node("P3", "3+ groups:\nChi-square test")

    p.edge("P_GROUPS", "P1", label="1")
    p.edge("P_GROUPS", "P2", label="2")
    p.edge("P_GROUPS", "P3", label="3+")

dot.edge("P_GATE", "P_GROUPS")


dot.node(
    "C_GATE",
    "Check conditions (chi-square)\n"
    "• Random/independent\n"
    "• Expected counts typically ≥ 5\n"
    "• If 2×2 and expected counts are small: Fisher exact"
)
dot.edge("H_CAT", "C_GATE")

dot.node(
    "CHI",
    "Categorical × Categorical:\n"
    "Chi-square test of independence\n"
    "(2×2 with small expected counts: Fisher exact)"
)
dot.edge("C_GATE", "CHI")


dot.node(
    "QQ_GATE",
    "Check conditions (relationship)\n"
    "• Random/independent\n"
    "• Roughly linear trend\n"
    "• No extreme outliers"
)
dot.edge("H_QQ", "QQ_GATE")

dot.node(
    "QQ",
    "Quant–Quant relationship:\n"
    "Correlation test (t test on r)\n"
    "or\n"
    "Regression slope test (t test on β₁)"
)
dot.edge("QQ_GATE", "QQ")

with dot.subgraph() as s2:
    s2.attr(rank="same")
    s2.node("M_GATE")
    s2.node("P_GATE")
    s2.node("C_GATE")
    s2.node("QQ_GATE")

out = dot.render("hypothesis_test_flowchart", cleanup=True)
print("Rendered:", out)
```
