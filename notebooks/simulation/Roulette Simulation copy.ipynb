# Roulette Simulator 
## Demos how often an "improbable" run/streak is likely to occur

```python
import random

casinos = 10
tables_per_casino = 20
hours = 24
spins_per_hour = 50
utilization = 0.85
k = 15              #streak length
seed = 1

random.seed(seed)

tables = casinos * tables_per_casino
spins_per_table = int(hours * spins_per_hour * utilization)

p_red = 18 / 38
p_black = 18 / 38   
def spin():
    r = random.random()
    if r < p_red:
        return "R"
    elif r < p_red + p_black:
        return "B"
    else:
        return "G"

total_streaks = 0

for _ in range(tables):
    run = 0
    last = None

    for _ in range(spins_per_table):
        s = spin()

        if s == "G":
            run = 0
            last = None
        elif s == last:
            run += 1
        else:
            run = 1
            last = s

        if run == k:
            total_streaks += 1

print("Casinos:", casinos)
print("Tables:", tables)
print("Spins per table:", spins_per_table)
print("Total spins:", tables * spins_per_table)
print(f"{k}-in-a-row streaks observed daily:", total_streaks)
```
