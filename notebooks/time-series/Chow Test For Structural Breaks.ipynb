# Chow Test For Structural Breaks

```python
#H0: No Structural Break
#HA: Structural Break at Breakpoint 
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
from scipy.stats import f

df = pd.read_csv('../Data/structural_break_data.csv')
df.head()
```

```python
#Define break point
break_point = 50

#Split data on break point
df1 = df[df['time'] < break_point]
df2 = df[df['time'] >= break_point]

X_full = sm.add_constant(df['time'])
y_full = df['value']

X1 = sm.add_constant(df1['time'])
y1 = df1['value']

X2 = sm.add_constant(df2['time'])
y2 = df2['value']

model_full = sm.OLS(y_full, X_full).fit()
model1 = sm.OLS(y1, X1).fit()
model2 = sm.OLS(y2, X2).fit()

#Calculate RSS
RSS_pooled = sum(model_full.resid ** 2)
RSS1 = sum(model1.resid ** 2)
RSS2 = sum(model2.resid ** 2)

#Degrees of freedom
k = X_full.shape[1]  #based on number of parameters (including constant)
n1 = len(y1)
n2 = len(y2)

#Chow F-statistic
F_chow = ((RSS_pooled - (RSS1 + RSS2)) / k) / ((RSS1 + RSS2) / (n1 + n2 - 2 * k))

df_num = k      #numerator degrees of freedom
df_den = n1 + n2 - 2 * k  #denominator degrees of freedom

p_value = 1 - f.cdf(F_chow, df_num, df_den)

print("Chow F-statistic:", F_chow)
print("p-value:", p_value)

#If p-value is less than alpha, reject null that there is no structural break / conclude there is a structural break
```
