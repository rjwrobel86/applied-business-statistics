# Multivariate Normal Distribution

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import multivariate_normal

#Define mean vector and covariance matrix
mu = np.array([170, 65])  
Sigma = np.array([[100, 50],  
                  [50, 40]])

samples = np.random.multivariate_normal(mu, Sigma, 1000)

x1_samples = samples[:, 0]
x2_samples = samples[:, 1]

empirical_mean = np.mean(samples, axis=0)
empirical_cov = np.cov(samples.T)

print("Empirical Mean:\n", empirical_mean)
print("Empirical Covariance Matrix:\n", empirical_cov)

sns.jointplot(x=x1_samples, y=x2_samples, kind="scatter", alpha=0.5)
plt.xlabel("Height (cm)")
plt.ylabel("Weight (kg)")
plt.show()

#Plot bivariate normal density
x, y = np.meshgrid(np.linspace(140, 200, 100), np.linspace(40, 90, 100))
pos = np.dstack((x, y))
rv = multivariate_normal(mu, Sigma)
z = rv.pdf(pos)

plt.contourf(x, y, z)
plt.xlabel("Height (cm)")
plt.ylabel("Weight (kg)")
plt.title("Bivariate Normal Distribution Density")
plt.colorbar(label="Density")
plt.show()
```
