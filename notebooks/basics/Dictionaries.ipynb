# Dictionaries 

```python
#Dictionaries

#Create a dictionary
demo_dict = {"name": "Bob", "age": 37, "city": "St. Louis"}
print(demo_dict)
```

```python
#Retrieve the value for a given key:

print(demo_dict.get("name"))
```

```python
#Get all the keys in the dictionary:

print(demo_dict.keys())  
```

```python
#Get all the values in the dictionary:

print(demo_dict.values())  
```

```python
#Get all key-value pairs

print(demo_dict.items())   
```

```python
#Check membership
print("name" in demo_dict)
print("salary" in demo_dict)
```

```python
#Add / update
demo_dict.update({"city":"chicago", "job":"secret agent"})
demo_dict
```

```python
#Remove
demo_dict.pop("job")
demo_dict
```

```python
#Dictionary comprehensions
upper_keys = {k.upper(): v for k, v in demo_dict.items()}
upper_keys
```
