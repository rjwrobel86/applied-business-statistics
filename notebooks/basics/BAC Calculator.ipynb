# Widmark BAC Calculator

```python
#A = alcohol per drink = 0.6 (beer, wine, or shot)
#D = # of drinks
#W = body weight in pounds
#r_m = male alcohol distribution ratio = 0.73
#r_w = female alcohol distribution ratio = 0.66
#H = hours since drinking began
#m = average alcohol metabolism rate per hour = 0.15
#c_1 = 5.14  = constant
#c_2 = 0.015 = constant 
```

```python
#How drunk are you?
A = 0.6 
#W = 250
#H = 1
r_m = 0.73
r_f = 0.66
m = 0.15
c_1 = 5.14
c_2 = 0.015
#D = 3

D = int(input("How many drinks?"))
W = int(input("Weight in lbs?"))
H = int(input("How long since started drinking?"))
G = (input("(M)ale or (F)emale?")).strip().upper()

if G == "M":
    r = r_m
elif G == "F":
    r = r_f

BAC = ((D*A*c_1)/(W*r)) - c_2*H
print(f"Your BAC is {BAC}")

if BAC > 0.08:
    print("You're drunk")
else:
    print("You're not drunk...yet")
```

```python
#How many drinks until drunk?
W = int(input("Weight in lbs? "))
H = float(input("How many hours since drinking started? "))
G = input("(M)ale or (F)emale? ").strip().upper()

if G == "M":
    r = r_m
elif G == "F":
    r = r_f
else:
    raise ValueError("Invalid gender input. Use 'M' or 'F'.")

#Target BAC
target_bac = 0.08

#Solve for number of drinks D:
D = ((target_bac + c_2 * H) * (W * r)) / (A * c_1)
D = round(D, 1)

print(f"It would take approximately {D} drinks to reach a BAC of {target_bac}")
```
