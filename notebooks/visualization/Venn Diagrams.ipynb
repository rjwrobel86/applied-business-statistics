```python
import matplotlib.pyplot as plt
from matplotlib_venn import venn2
```

```python
plt.figure(figsize=(8, 6))
venn = venn2(subsets=(1.0, 1.0, 0.5), set_labels=('A: Bank Teller', 'B: Political'))
venn.get_label_by_id('10').set_text('P(A)\nBank Teller')
venn.get_label_by_id('01').set_text('P(B)\nPolitical')
venn.get_label_by_id('11').set_text('P(A ∩ B)\nBank Teller & Political')

plt.title("Venn Diagram: P(A ∩ B) ≤ P(A)")
plt.show()
```
