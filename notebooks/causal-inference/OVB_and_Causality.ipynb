```python
import pandas as pd
import statsmodels.api as sms

df = pd.read_csv("../Data/Ice_Cream_Kills.csv")
df.head()
```

```python
X1 = sms.add_constant(df["icecream_consumption"])
m1 = sms.OLS(df["drownings"], X1).fit()
m1.summary()
```

```python
X2 = sms.add_constant(df[["icecream_consumption", "avg_temp_F"]])
m2 = sms.OLS(df["drownings"], X2).fit()
m2.summary()
```
