```python
import numpy as np
import pandas as pd
import statsmodels.api as sm

df = pd.read_csv('../Data/card_education_sim.csv')
df.head()
```

```python
#First Stage: regress education on distance and parent_income
X_first_stage = sm.add_constant(df[["distance", "parent_income"]])
first_stage = sm.OLS(df["education"], X_first_stage).fit()
df["education_hat"] = first_stage.predict(X_first_stage)
```

```python
#Second Stage: regress wage on predicted education and parent_income
X_second_stage = sm.add_constant(df[["education_hat", "parent_income"]])
second_stage = sm.OLS(df["wage"], X_second_stage).fit()
second_stage.summary()
```

```python
#OLS Comparison
X_ols = sm.add_constant(df[["education", "parent_income"]])
ols = sm.OLS(df["wage"], X_ols).fit()
ols.summary()
```
