```python
import numpy as np
import pandas as pd
from semopy import Model, semplot

np.random.seed(0)
n = 300
education = np.random.normal(14, 2, n)
income = 10 + 2.5 * education + np.random.normal(0, 5, n)
job_satisfaction = 1 + 0.4 * income + 0.3 * education + np.random.normal(0, 4, n)

df = pd.DataFrame({
    "Education": education,
    "Income": income,
    "JobSatisfaction": job_satisfaction
})

#Define path model
desc = """
Income ~ Education
JobSatisfaction ~ Income + Education
"""

#Fit the model
model = Model(desc)
model.fit(df)

#Inspect output and check structure
estimates = model.inspect()
print("\nModel Estimates:\n", estimates[['lval', 'op', 'rval', 'Estimate']])

#Extract effects using correct column names
def get_estimate(est_df, lhs, rhs):
    match = est_df[(est_df["lval"] == lhs) & (est_df["rval"] == rhs)]
    return match["Estimate"].values[0] if not match.empty else np.nan

b1 = get_estimate(estimates, 'Income', 'Education')            #Education → Income
b2 = get_estimate(estimates, 'JobSatisfaction', 'Income')      #Income → JobSatisfaction
b3 = get_estimate(estimates, 'JobSatisfaction', 'Education')   #Education → JobSatisfaction

#Compute indirect and total effects
indirect = b1 * b2
total = b3 + indirect

print(f"\nDirect effect of Education on Income: {b1:.3f}")
print(f"Direct effect of Income on Job Satisfaction: {b2:.3f}")
print(f"Direct effect of Education on Job Satisfaction: {b3:.3f}")
print(f"Indirect effect (Education → Income → Satisfaction): {indirect:.3f}")
print(f"Total effect of Education on Satisfaction: {total:.3f}")

#Path diagram
semplot(model, "sem_path_diagram.png", show=True)
```
